```python
# Creating a tabular list of the neighborhoods of Bangalore which shall be considered for this project
link = 'https://en.wikipedia.org/wiki/List_of_neighbourhoods_in_Bangalore'
html_data = requests.get(link).text
table = pd.read_html(link)[0]
table_data_0 = pd.DataFrame(table)
table_data_0 = table_data_0.drop(['Image','Summary'], axis =1)
table_data_0
table = pd.read_html(link)[1]
table_data_1 = pd.DataFrame(table)
table_data_1 = table_data_1.drop(['Image','Summary'], axis =1)
table_data_1
table = pd.read_html(link)[2]
table_data_2 = pd.DataFrame(table)
table_data_2 = table_data_2.drop(['Image','Summary'], axis =1)
table_data_2
table = pd.read_html(link)[3]
table_data_3 = pd.DataFrame(table)
table_data_3 = table_data_3.drop(['Image','Summary'], axis =1)
table_data_3
table_data_3
table = pd.read_html(link)[4]
table_data_4 = pd.DataFrame(table)
table_data_4 = table_data_4.drop(['Image','Summary'], axis =1)
table_data_4
table_data_4
table_data = pd.concat([table_data_0,table_data_1,table_data_2,table_data_3,table_data_4], axis = 0)
table_data = table_data.reset_index()
table_data = table_data.drop(['index'], axis = 1)
table_data = table_data.drop([3],axis=0)
table_data['Name'] = table_data['Name']+',Bangalore'
table_data
```

```python
#Finding and adding Latitudes and Longitudes of the neighborhoods in the above Dataframe
Nbr = table_data['Name'].tolist()
print(Nbr)
geolocator = Nominatim(user_agent="mygeocoder")
coords = []
for Nbr in Nbr:
    location = geolocator.geocode(Nbr.format(Nbr))
    coords.append([Nbr,location.latitude,location.longitude])
    print(Nbr,location.latitude,location.longitude)
```

```python
# A neat dataframe is thus created..
df = pd.DataFrame(coords, columns=["Name", "Latitude","Longitude"])
df = df.drop([36], axis=0)
df
```
